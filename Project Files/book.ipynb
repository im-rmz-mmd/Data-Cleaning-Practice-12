```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Book.csv')
df
```

```python
df = df.drop(columns = {'Segment', 'Month Number', 'Month Name', 'Year', 'Units Sold', 'Gross Sales', 'COGS'})
df = df.drop_duplicates()
df = df.dropna()
df
```

```python
df = df.rename(columns = {' Sales' : 'Sales'})
df
```

```python
df['Profit'] = df['Profit'].abs()
df
```

```python
df['Date'] = pd.to_datetime(df['Date'], format = '%d-%b-%y')
df.dtypes
```

```python
df.to_excel('Cleaned_book.xlsx', index = False)
```
